#NLP

There are a wide range of problems that can be solved using NLP,  including, but not limited to, speech recognition, sentiment analysis, topic modelling etc. NLP has many real-life applications. E.g., Alexa, Google Assistant etc.


Let us begin by understanding a simple text pre-processing pipeline.

```python
# We will assign a variable a text and ways to extract words from the statement.

# split into words by white space
text = "Deepak gave us a real chance. Quite an exciting game, disappointing that we ended up on the losing side. it's quite obvious where we've gone wrong and it's for everyone to see, our shot selection has been really poor as batsmen. Even as bowlers we weren't hitting the good areas consistently. We didn't pressure the opposition for long periods so we ended up losing the series. We need to recognise now how we can get better. what has always stood out for us is the energy and the passion, and the skill and understanding of the game. These kind of mistakes will keep happening, but it's. important we learn and don't keep repeating the mistakes. In the ODIs we repeated some of the mistakes. We need to look at ourselves in the mirror and have some hard conversations. We've had a great time in South Africa, we were looked after really well, thanks to CSA. We had an exciting series, we showed a lot of fight, in patches. We'll go back with learnings and memories, and get better for sure."
word = text.split()
print(word[:30])

# We use regex to split a word like 'pre-processed' into 2 word 'pre' and 'processed'
import re
word = re.split(r'\W+', text)
print(word[:15])
```

# Text Pre-processing

```python
# split into words
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
tokens = word_tokenize(text)


# convert to lower case
tokens = [w.lower() for w in tokens]
#print(tokens[:10])

# replace punctuation from each word
import nltk
import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in tokens]
#print(stripped)

# remove remaining tokens that are not alphabetic
words = [word for word in stripped if word.isalpha()]
print(words)

# filter out stop words
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
#print(stop_words)
words = [w for w in words if not w in stop_words]
print(words[:25])
```

```python
#Part of Specch Tagging also known as POS tags. In this section, we will tag each word into its POS tag
#Abbrevation mapping https://www.guru99.com/pos-tagging-chunking-nltk.html

import nltk
nltk.download('averaged_perceptron_tagger')
print(words)

tagged = nltk.pos_tag(words)
print(tagged[0:25])
```

# Stemming and Lemmatization

```python
#Stemming is technique to remove suffixes from the words
#Eg 'openly' becomes 'open' , 'excites' becomes 'excite'. 

import nltk
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
print(stemmed[:10])

#Stemming can be a challenge, as it can under stem or over stem
#sent1 = "Programmers program with programming languages"
#sent2 = word_tokenize(sent1)
#for w in sent2:
#    print(w, " : ", porter.stem(w))

#As you can see in stemming, we haven't got normalized words. To make it better we use Lemmitize. Lemmatization 
#extracts the root word or retain the meaning and tense of the words in the context to the words.

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmed = [lemmatizer.lemmatize(word) for word in words]
print(lemmed[:10])
```

```python
#Here we create a term document matrix for easy indexing and searching of terms and basic statistical values.
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

docs = ['Rohan is is is a data scientist', 'Data science is a great career career', 'You need is is math to understand data']
vec = CountVectorizer()


X = vec.fit_transform(docs)
df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df)
```

# TF-iDF Technique

TF-IDF is a weighted model commonly used for information retrieval problems. It aims to convert the text documents into vector models on the basis of occurrence of words in the documents without taking considering the exact ordering

Term Frequency (TF)  TF for a term t is defined as the count of a term t in a document D

Inverse Document Frequency (IDF)  IDF for a term is defined as logarithm of ratio of total documents available in the corpus and number of documents containing the term T.

#Text Classification

Text classification is a machine learning technique that assigns a set of predefined categories to open-ended text.

Text classification is one of the fundamental tasks in natural language processing with broad applications such as sentiment analysis, topic labeling, spam detection, and intent detection.

For e.g. Gmail classifies a mail as primary, social, updates, promotions and forum based on sender, content, image, etc.

```python

from textblob.classifiers import NaiveBayesClassifier as NBC
from textblob import TextBlob
training_corpus = [
                   ('I am really exhausted with similar work.', 'Negative'),
                   ("I can't operate with the key", 'Negative'),
                   ('He is my worst enemy ever!', 'Negative'),
                   ('My management style is poor and need to improve.', 'Negative'),
                   ('I love Italian food, especially Pizza.', 'Positive'),
                   ('This is an brilliant place!', 'Positive'),
                   ('I feel very good about the environment am in.', 'Positive'),
                   ('This is my best work at Ingram Micro.', 'Positive'),
                   ("What an awesome view from my room", 'Positive'),
                   ('I do not like this cycle', 'Negative')]
test_corpus = [
                ("I am not feeling great.", 'Negative'), 
                ("I feel ecstattic!", 'Positive'), 
                ('John is a friend of mine.', 'postive'), 
                ("I can't believe I'm doing this.", 'Negative'), 
                ('The date was good.', 'Positive'), 
                ('I do not enjoy my job', 'Negative')]


model = NBC(training_corpus) 
print(model.accuracy(test_corpus))

print(model.classify("Deepak gave us a real chance. Quite an exciting game, disappointing that we ended up on the losing side. it's quite obvious where we've gone wrong and it's for everyone to see, our shot selection has been really poor as batsmen. Even as bowlers we weren't hitting the good areas consistently. We didn't pressure the opposition for long periods so we ended up losing the series. We need to recognise now how we can get better. what has always stood out for us is the energy and the passion, and the skill and understanding of the game. These kind of mistakes will keep happening, but it's. important we learn and don't keep repeating the mistakes. In the ODIs we repeated some of the mistakes. We need to look at ourselves in the mirror and have some hard conversations. We've had a great time in South Africa, we were looked after really well, thanks to CSA. We had an exciting series, we showed a lot of fight, in patches. We'll go back with learnings and memories, and get better for sure."))

print(model.classify("            "))
```
